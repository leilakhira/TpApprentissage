```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import os
import time
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.utils import shuffle
X, y = shuffle(dataset_feature, dataset_labels, random_state=1)
# only use the first N samples to limit training time
num_samples = int(len(X)*0.1)
X_, y_ = X[:num_samples], y[:num_samples]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()

# Adapter le scaler aux données et le transformer
X_scaled = scaler.fit_transform(X_)
```

```python

#split into train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_,test_size=0.2)
print("X_train.shape:", X_train.shape)
print("y_train.shape:", y_train.shape)
print("X_test.shape:", X_test.shape)
print("y_test.shape:", y_test.shape)
```

```python
# Créer le modèle SVM
svm_model = svm.SVC(kernel='linear')   

# Appliquer la validation croisée
cross_val_scores_svm = cross_val_score(svm_model, X_train,y_train, cv=5)  # cv représente le nombre de plis

# Afficher les résultats de la validation croisée
print("Moyenne des scores de validation croisée SVM :", np.mean(cross_val_scores_svm))
```

```python
# Créer le modèle Random Forest
rf_model = RandomForestClassifier()  
# Appliquer la validation croisée
cross_val_scores_rf = cross_val_score(rf_model, X_train,y_train, cv=5)

# Afficher les résultats de la validation croisée
print("Moyenne des scores de validation croisée Random Forest :", np.mean(cross_val_scores_rf))
```

```python
# Créer le modèle AdaBoost
adaboost_model = AdaBoostClassifier() 

# Appliquer la validation croisée
cross_val_scores_adaboost = cross_val_score(adaboost_model,X_train,y_train, cv=5)

# Afficher les résultats de la validation croisée
print("Moyenne des scores de validation croisée AdaBoost :", np.mean(cross_val_scores_adaboost))
```

```python
# Créer le modèle Gradient Boosting
gradientboost_model = GradientBoostingClassifier()

# Appliquer la validation croisée
cross_val_scores_gradientboost = cross_val_score(gradientboost_model,X_train,y_train, cv=5)

# Afficher les résultats de la validation croisée
print("Moyenne des scores de validation croisée Gradient Boosting :", np.mean(cross_val_scores_gradientboost))
```

```python
svm_model = SVC()
rf_model = RandomForestClassifier() 
adaboost_model = AdaBoostClassifier()
gradientboost_model = GradientBoostingClassifier()
accuracies =[]
models=[svm_model,rf_model,adaboost_model,gradientboost_model]
def evaluation_metrics(model):
    
    # Entraîner le modèle
    model.fit(X_train, y_train)
    
    # Faire des prédictions sur l'ensemble de test
    y_pred = model.predict(X_test)

    # Évaluer l'accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print("Accuracy du modèle :", accuracy)

    # Afficher la confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix :\n", conf_matrix)
    return accuracies 

    
for model in models:
    print("Evaluation de la qualité du modèle d’apprentissage :",model)
    evaluation_metrics(model)
    print("------------------------------------------------------------------------------------")
    
```

```python
# Affichage du diagramme d'accuracy
model_names = ['SVM', 'Random Forest', 'AdaBoost', 'Gradient Boosting']
x_pos = np.arange(len(model_names))

plt.bar(x_pos, accuracies, align='center', alpha=0.7)
plt.xticks(x_pos, model_names)
plt.ylabel('Accuracy')
plt.title('Accuracy de différents modèles')
plt.show()
```

```python
# Définir la grille des hyperparamètres à rechercher
models=[svm_model,rf_model,adaboost_model,gradientboost_model]
svm_param_grid = {'C': [0.1, 1, 10, 100],
                  'kernel': ['rbf']}

rf_param_grid = {'n_estimators': [50, 100, 200],
                 'max_depth': [10, 20, 30],
                 'min_samples_split': [2, 5, 10],
                 'min_samples_leaf': [1, 2, 4]}

adaboost_param_grid = {'n_estimators': [50, 100, 200],
                      'learning_rate': [0.01, 0.1, 1.0],
                      'base_estimator': [None, DecisionTreeClassifier(max_depth=1)]}

gradientboost_param_grid = {'n_estimators': [50, 100, 200],
                           'learning_rate': [0.01, 0.1, 0.2],
                           'max_depth': [3, 5, 7],
                           'min_samples_split': [2, 5, 10]}
def model_grid_search(model,grid_search_parm):
    grid_search = GridSearchCV(model, grid_search_parm, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    print(grid_search.best_params_, grid_search.best_score_)
    

#for grid in grid_sreachs:
print("les grilles d'hyperparamètres du modèle d’apprentissage :SVM model")
model_grid_search(svm_model,svm_param_grid)
    
print("les grilles d'hyperparamètres du modèle d’apprentissage : RandomForest model")
model_grid_search(rf_model,rf_param_grid)
    
print("les grilles d'hyperparamètres du modèle d’apprentissage : AdaBoost model")
model_grid_search(adaboost_model,adaboost_param_grid)
    
print("les grilles d'hyperparamètres du modèle d’apprentissage :GradientBoosting model")
model_grid_search(gradientboost_model,gradientboost_param_grid)
```

```python
from sklearn.inspection import permutation_importance
def display_feat_imp(model):
    # Utilisation du  permutation_importance pour évaluer l'importance des features
    result = permutation_importance(model, X_test, y_test, n_repeats=30, random_state=42)
    # Récupérez les scores d'importance
    importance_scores = result.importances_mean
    feature_names = ['AGEP', 'COW', 'SCHL', 'MAR', 'OCCP', 'POBP', 'RELP', 'WKHP', 'SEX','RAC1P']
    df_featimp = pd.DataFrame(importance_scores, columns=["Feature Importance"])
    df_featimp["Feature Name"] = feature_names
    df_featimp = df_featimp.sort_values(by="Feature Importance", ascending=False)
    #print(df_featimp)
    df_featimp.plot.barh(y="Feature Importance", x="Feature Name", title="Feature importance pour le model"+ str(model),color="red")
    
```

```python
for model in models:
    display_feat_imp(model)
```
